```python
# Q.1.1.(b):-Create a function to find the median of a list of numbers.
#ans:-
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2

    if n % 2 == 0:  # Even number of elements
        median_value = (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:  # Odd number of elements
        median_value = sorted_numbers[mid]

    return median_value

# Example usage:
median_value = calculate_median(int_list)
print(f"Median: {median_value}")
```

```python
#Q.1.(2):- Develop a program to compute the mode of a list of integers.
#Ans:-
from collections import Counter

def calculate_mode(numbers):
    # Count the frequency of each number in the list
    frequency = Counter(numbers)
    
    # Find the maximum frequency
    max_frequency = max(frequency.values())
    
    # Find the number(s) with the maximum frequency (the mode)
    mode = [number for number, freq in frequency.items() if freq == max_frequency]
    
    # If there's only one mode, return it as a single value, otherwise return the list of modes
    return mode if len(mode) > 1 else mode[0]

# Example usage:
mode_value = calculate_mode(int_list)
print(f"Mode: {mode_value}")
```

```python
#Q.1.(3):- Implement a function to calculate the weighted mean of list of values and their corresponding weights.
#ans:-
def calculate_weighted_mean(values, weights):
    if len(values) != len(weights):
        raise ValueError("The length of values and weights must be the same.")

    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)

    return weighted_sum / total_weight

# Example usage:
values = [90, 100, 110, 120, 130]
weights = [1, 2, 3, 4, 5]
weighted_mean_value = calculate_weighted_mean(values, weights)
print(f"Weighted Mean: {weighted_mean_value}")
```

```python
#Q.1.(4):- Write a Python function to find the geometric mean of a list of positive numbers.
#Ans:-
import math

def calculate_geometric_mean(numbers):
    # Ensure all numbers are positive
    if any(n <= 0 for n in numbers):
        raise ValueError("All numbers must be positive to calculate the geometric mean.")
    
    product = math.prod(numbers)
    n = len(numbers)
    
    # Calculate the nth root of the product
    geometric_mean = product ** (1/n)
    
    return geometric_mean

# Example usage:
numbers = [90, 100, 110, 120, 130]
geometric_mean_value = calculate_geometric_mean(numbers)
print(f"Geometric Mean: {geometric_mean_value}")
```

```python
# Q.1.(7):- Implement a Python program to find the trimmed mean of a list,excluding a certain percentage of outliers.
#Ans:-
def calculate_trimmed_mean(numbers, trim_percent):
    if len(numbers) == 0:
        raise ValueError("The list of numbers must not be empty.")
    if not (0 <= trim_percent < 50):
        raise ValueError("Trim percentage must be between 0 and 50.")
    
    # Sort the numbers
    sorted_numbers = sorted(numbers)
    
    # Calculate the number of values to trim from each end
    trim_count = int(len(sorted_numbers) * trim_percent / 100)
    
    # Trim the numbers
    trimmed_numbers = sorted_numbers[trim_count:-trim_count]
    
    # Check if there are still numbers left after trimming
    if len(trimmed_numbers) == 0:
        raise ValueError("After trimming, no numbers left to calculate the mean.")
    
    # Calculate the mean of the trimmed numbers
    trimmed_mean = sum(trimmed_numbers) / len(trimmed_numbers)
    
    return trimmed_mean

# Example usage
numbers = [90, 100, 110, 120, 130]
trim_percent = 20  # Percentage of values to trim from each end

# Calculate the trimmed mean
trimmed_mean_value = calculate_trimmed_mean(numbers, trim_percent)

# Print the result
print(f"Trimmed Mean: {trimmed_mean_value}")
```

```python
#Q.2.(3):- Create a program to find the varience and standard deviation of a list of numbers.
#Ans:-
import math

def calculate_variance(numbers):
    if len(numbers) == 0:
        raise ValueError("The list of numbers must not be empty.")
    
    mean = sum(numbers) / len(numbers)
    variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
    
    return variance

def calculate_standard_deviation(numbers):
    variance = calculate_variance(numbers)
    return math.sqrt(variance)

# Example usage with int_list2
import random

# Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Calculate variance and standard deviation
variance_value = calculate_variance(int_list2)
standard_deviation_value = calculate_standard_deviation(int_list2)

# Print the results
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {standard_deviation_value}")
```

```python
# Q.2.(5):- Build a program to calculate the coeffient of variation for a dataset.
#Ans:-
import numpy as np

def calculate_mean(numbers):
    return np.mean(numbers)

def calculate_standard_deviation(numbers):
    return np.std(numbers, ddof=0)  # Use ddof=0 for population standard deviation

def calculate_coefficient_of_variation(numbers):
    if len(numbers) == 0:
        raise ValueError("The list of numbers must not be empty.")
    
    mean = calculate_mean(numbers)
    std_dev = calculate_standard_deviation(numbers)
    
    if mean == 0:
        raise ValueError("Mean of the dataset is zero, CV cannot be computed.")
    
    # Calculate the Coefficient of Variation
    cv = (std_dev / mean) * 100
    
    return cv

# Example usage with int_list2
import random

# Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Calculate the Coefficient of Variation
cv_value = calculate_coefficient_of_variation(int_list2)

# Print the result
print(f"Coefficient of Variation (CV): {cv_value:.2f}%")
```

```python
#Q.2.(6):- Write a Python function to find the mean absolute deviation(MAD) of a list of numbers.
#Ans:-
import numpy as np

def calculate_mean_absolute_deviation(numbers):
    if len(numbers) == 0:
        raise ValueError("The list of numbers must not be empty.")
    
    # Calculate the mean of the numbers
    mean = np.mean(numbers)
    
    # Calculate the absolute deviations from the mean
    absolute_deviations = [abs(x - mean) for x in numbers]
    
    # Calculate the Mean Absolute Deviation (MAD)
    mad = np.mean(absolute_deviations)
    
    return mad

# Example usage with int_list2
import random

# Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Calculate the Mean Absolute Deviation (MAD)
mad_value = calculate_mean_absolute_deviation(int_list2)

# Print the result
print(f"Mean Absolute Deviation (MAD): {mad_value:.2f}")
```

```python
#Q.2.(8):- Implement a function to find the range-based coeffient of dispersion for a dataset
#Ans:-
import numpy as np

def calculate_range_based_cd(numbers):
    if len(numbers) == 0:
        raise ValueError("The list of numbers must not be empty.")
    
    # Calculate the mean of the numbers
    mean = np.mean(numbers)
    
    if mean == 0:
        raise ValueError("Mean of the dataset is zero, CD cannot be computed.")
    
    # Calculate the range of the numbers
    range_value = np.max(numbers) - np.min(numbers)
    
    # Calculate the Coefficient of Dispersion (CD)
    cd = range_value / mean
    
    return cd

# Example usage with int_list2
import random

# Generate a list of 500 random integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Calculate the Range-Based Coefficient of Dispersion (CD)
cd_value = calculate_range_based_cd(int_list2)

# Print the result
print(f"Range-Based Coefficient of Dispersion (CD): {cd_value:.2f}")
```

```python
#Q.3.Write a Python class representing a discreate random variable with methods to calculate its expected value and varience.
#Ans:-
class DiscreteRandomVariable:
    def __init__(self, probabilities, values):
        if len(probabilities) != len(values):
            raise ValueError("Length of probabilities and values must be the same.")
        if not (0.99 <= sum(probabilities) <= 1.01):
            raise ValueError("The sum of probabilities must be approximately 1.")
        if any(p < 0 for p in probabilities):
            raise ValueError("Probabilities must be non-negative.")
        
        self.probabilities = probabilities
        self.values = values

    def expected_value(self):
        # Calculate the expected value (mean)
        return sum(p * v for p, v in zip(self.probabilities, self.values))

    def variance(self):
        # Calculate the expected value
        mean = self.expected_value()
        
        # Calculate the variance
        return sum(p * (v - mean) ** 2 for p, v in zip(self.probabilities, self.values))

# Example usage
probabilities = [0.2, 0.5, 0.3]  # Probabilities associated with each value
values = [1, 2, 3]  # Discrete values of the random variable

# Create an instance of DiscreteRandomVariable
random_variable = DiscreteRandomVariable(probabilities, values)

# Calculate the expected value and variance
expected_value = random_variable.expected_value()
variance = random_variable.variance()

# Print the results
print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")
```

```python
#Q.4:- Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.
#Ans:-
class FairSixSidedDie:
    def __init__(self):
        # Define possible outcomes and their probabilities
        self.values = [1, 2, 3, 4, 5, 6]
        self.probabilities = [1/6] * 6  # Equal probability for each face of the die

    def expected_value(self):
        # Calculate the expected value (mean)
        return sum(p * v for p, v in zip(self.probabilities, self.values))

    def variance(self):
        # Calculate the expected value
        mean = self.expected_value()
        
        # Calculate the variance
        return sum(p * (v - mean) ** 2 for p, v in zip(self.probabilities, self.values))

# Create an instance of FairSixSidedDie
die = FairSixSidedDie()

# Calculate the expected value and variance
expected_value = die.expected_value()
variance = die.variance()

# Print the results
print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")
```

```python
#Q.5:- Create a Python function to generate random samples from a given probability distribution(e.g.,binomial,Poisson) and calculate their mean and variance
#Ans:-
import numpy as np

def generate_samples_and_calculate_stats(distribution, *params, size=1000):
    """
    Generate random samples from a given probability distribution and calculate their mean and variance.

    Parameters:
    - distribution (str): Type of probability distribution ('binomial' or 'poisson').
    - *params: Parameters for the distribution (e.g., n, p for binomial; lambda for Poisson).
    - size (int): Number of random samples to generate.

    Returns:
    - mean (float): Mean of the generated samples.
    - variance (float): Variance of the generated samples.
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")
    
    # Calculate mean and variance
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example usage for binomial distribution
binomial_params = (10, 0.5)  # n=10, p=0.5
mean_binomial, variance_binomial = generate_samples_and_calculate_stats('binomial', *binomial_params)

print(f"Binomial Distribution - Mean: {mean_binomial:.2f}, Variance: {variance_binomial:.2f}")

# Example usage for Poisson distribution
poisson_param = 5  # lambda=5
mean_poisson, variance_poisson = generate_samples_and_calculate_stats('poisson', poisson_param)

print(f"Poisson Distribution - Mean: {mean_poisson:.2f}, Variance: {variance_poisson:.2f}")
```

```python
#Q.6:-Write a Python script to generate random numbers from a Gaussian(normal) distribution and compute the mean,variance,and standard deviation of the samples.
#Ans:-
import numpy as np

def generate_normal_samples(mean=0, std_dev=1, size=1000):
    """
    Generate random samples from a Gaussian (normal) distribution and compute their mean, variance, and standard deviation.

    Parameters:
    - mean (float): Mean of the normal distribution.
    - std_dev (float): Standard deviation of the normal distribution.
    - size (int): Number of random samples to generate.

    Returns:
    - mean (float): Mean of the generated samples.
    - variance (float): Variance of the generated samples.
    - std_deviation (float): Standard deviation of the generated samples.
    """
    # Generate random samples from a normal distribution
    samples = np.random.normal(mean, std_dev, size)
    
    # Calculate mean, variance, and standard deviation
    mean_value = np.mean(samples)
    variance_value = np.var(samples)
    std_deviation_value = np.std(samples)
    
    return mean_value, variance_value, std_deviation_value

# Example usage
mean = 0         # Mean of the normal distribution
std_dev = 1      # Standard deviation of the normal distribution
size = 1000      # Number of samples

# Generate samples and compute statistics
mean_value, variance_value, std_deviation_value = generate_normal_samples(mean, std_dev, size)

# Print the results
print(f"Mean of samples: {mean_value:.2f}")
print(f"Variance of samples: {variance_value:.2f}")
print(f"Standard Deviation of samples: {std_deviation_value:.2f}")
```

```python
#Q.7.(2):- Create a program that determines whether the columns exhibit positive skewness,negative skewness,or is approximately symmetric.
#Ans:-
import seaborn as sns
import pandas as pd
from scipy.stats import skew

def calculate_skewness(dataframe, column):
    """
    Calculate the skewness of a specified column in a DataFrame.

    Parameters:
    - dataframe (pd.DataFrame): The DataFrame containing the data.
    - column (str): The name of the column for which to calculate skewness.

    Returns:
    - skewness (float): The skewness of the column.
    """
    if column not in dataframe.columns:
        raise ValueError(f"Column '{column}' does not exist in the DataFrame.")
    
    # Calculate skewness
    skewness = skew(dataframe[column].dropna())
    
    return skewness

def determine_skewness_type(skewness):
    """
    Determine the skewness type based on the skewness value.

    Parameters:
    - skewness (float): The skewness value.

    Returns:
    - skewness_type (str): Type of skewness ('positive', 'negative', or 'symmetric').
    """
    if skewness > 0:
        return 'positive'
    elif skewness < 0:
        return 'negative'
    else:
        return 'symmetric'

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Calculate skewness for 'total_bill' and 'tip'
skewness_total_bill = calculate_skewness(tips, 'total_bill')
skewness_tip = calculate_skewness(tips, 'tip')

# Determine skewness type
skewness_type_total_bill = determine_skewness_type(skewness_total_bill)
skewness_type_tip = determine_skewness_type(skewness_tip)

# Print results
print(f"Skewness of 'total_bill': {skewness_total_bill:.2f} ({skewness_type_total_bill})")
print(f"Skewness of 'tip': {skewness_tip:.2f} ({skewness_type_tip})")
```

```python
#Q.7.(3):- Write a function that calculates the covariance between two columns.
#Ans:-
import pandas as pd

def calculate_covariance(dataframe, column1, column2):
    """
    Calculate the covariance between two specified columns in a DataFrame.

    Parameters:
    - dataframe (pd.DataFrame): The DataFrame containing the data.
    - column1 (str): The name of the first column.
    - column2 (str): The name of the second column.

    Returns:
    - covariance (float): The covariance between the two columns.
    """
    if column1 not in dataframe.columns or column2 not in dataframe.columns:
        raise ValueError(f"One or both columns '{column1}' and '{column2}' do not exist in the DataFrame.")
    
    # Drop rows with missing values in either column
    data = dataframe[[column1, column2]].dropna()
    
    # Calculate means of the columns
    mean1 = data[column1].mean()
    mean2 = data[column2].mean()
    
    # Calculate covariance
    covariance = ((data[column1] - mean1) * (data[column2] - mean2)).mean()
    
    return covariance

# Example usage with 'tips' dataset
import seaborn as sns

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Calculate covariance between 'total_bill' and 'tip'
covariance_total_bill_tip = calculate_covariance(tips, 'total_bill', 'tip')

# Print the result
print(f"Covariance between 'total_bill' and 'tip': {covariance_total_bill_tip:.2f}")
```

```python
#Q.7.(4):- Implement a Python program that calculates the Pearson correlation coefficient between two columns.
#Ans:-
import pandas as pd
import numpy as np

def calculate_pearson_correlation(dataframe, column1, column2):
    """
    Calculate the Pearson correlation coefficient between two specified columns in a DataFrame.

    Parameters:
    - dataframe (pd.DataFrame): The DataFrame containing the data.
    - column1 (str): The name of the first column.
    - column2 (str): The name of the second column.

    Returns:
    - correlation (float): The Pearson correlation coefficient between the two columns.
    """
    if column1 not in dataframe.columns or column2 not in dataframe.columns:
        raise ValueError(f"One or both columns '{column1}' and '{column2}' do not exist in the DataFrame.")
    
    # Drop rows with missing values in either column
    data = dataframe[[column1, column2]].dropna()
    
    # Calculate means of the columns
    mean1 = data[column1].mean()
    mean2 = data[column2].mean()
    
    # Calculate covariance
    covariance = ((data[column1] - mean1) * (data[column2] - mean2)).mean()
    
    # Calculate standard deviations
    std_dev1 = data[column1].std()
    std_dev2 = data[column2].std()
    
    # Calculate Pearson correlation coefficient
    if std_dev1 == 0 or std_dev2 == 0:
        raise ValueError("Standard deviation of one of the columns is zero; correlation is undefined.")
    
    correlation = covariance / (std_dev1 * std_dev2)
    
    return correlation

# Example usage with 'tips' dataset
import seaborn as sns

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Calculate Pearson correlation coefficient between 'total_bill' and 'tip'
pearson_correlation_total_bill_tip = calculate_pearson_correlation(tips, 'total_bill', 'tip')

# Print the result
print(f"Pearson correlation coefficient between 'total_bill' and 'tip': {pearson_correlation_total_bill_tip:.2f}")
```

```python
#Q.8:- write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.
#Ans:-
import numpy as np
from scipy.stats import norm

def calculate_pdf(x, mean=0, std_dev=1):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given value.

    Parameters:
    - x (float or np.ndarray): The value(s) at which to evaluate the PDF.
    - mean (float): The mean of the normal distribution.
    - std_dev (float): The standard deviation of the normal distribution.

    Returns:
    - pdf (float or np.ndarray): The PDF value(s) at the given value(s).
    """
    # Calculate PDF using scipy.stats.norm
    pdf = norm.pdf(x, loc=mean, scale=std_dev)
    
    return pdf

# Example usage
x_values = np.linspace(-5, 5, 100)  # Values at which to evaluate the PDF
mean = 0
std_dev = 1

# Calculate PDF for the given values
pdf_values = calculate_pdf(x_values, mean, std_dev)

# Print some of the results
print(f"PDF at x = {x_values[0]:.2f}: {pdf_values[0]:.4f}")
print(f"PDF at x = {x_values[50]:.2f}: {pdf_values[50]:.4f}")
print(f"PDF at x = {x_values[-1]:.2f}: {pdf_values[-1]:.4f}")
```

```python
#Q.10:-Write a Python function to calculate the probability mass function(PMF) of Poisson distribution.
#Ans:-
from scipy.stats import poisson

def calculate_poisson_pmf(k, mu):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.

    Parameters:
    - k (int or np.ndarray): The number of occurrences (must be a non-negative integer).
    - mu (float): The average rate (mean) of occurrences ().

    Returns:
    - pmf (float or np.ndarray): The PMF value(s) for the given number of occurrences.
    """
    if mu <= 0:
        raise ValueError("Mean (lambda) must be positive.")
    
    # Calculate PMF using scipy.stats.poisson
    pmf = poisson.pmf(k, mu)
    
    return pmf

# Example usage
k_values = [0, 1, 2, 3, 4, 5]  # Number of occurrences
mu = 3  # Mean of the Poisson distribution

# Calculate PMF for the given number of occurrences
pmf_values = calculate_poisson_pmf(k_values, mu)

# Print the results
for k, pmf in zip(k_values, pmf_values):
    print(f"PMF for k = {k}: {pmf:.4f}")
```

```python
#Q.11:- A company wants to test if a new website layout leads to a higher coversion rate(percentage of visitors who make a purchase).They collect data from the old and new layouts to compare.
# To generate the data use the following command:
''''Python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1]*50+[0]*950)
# 70 purchases out of 1000 visitors
new_layout = np.array([1]*70+[0]*930)
'''''
#    Apply z-test to find which layout is successfully.
#Ans:-
import numpy as np
from scipy.stats import norm

# Generate data
old_layout = np.array([1]*50 + [0]*950)
new_layout = np.array([1]*70 + [0]*930)

# Calculate proportions
p1 = np.mean(old_layout)  # Proportion for old layout
p2 = np.mean(new_layout)  # Proportion for new layout

# Sample sizes
n1 = len(old_layout)
n2 = len(new_layout)

# Pooled proportion
p_pooled = (np.sum(old_layout) + np.sum(new_layout)) / (n1 + n2)

# Standard error
se = np.sqrt(p_pooled * (1 - p_pooled) * (1/n1 + 1/n2))

# Z-statistic
z = (p1 - p2) / se

# P-value
p_value = 2 * (1 - norm.cdf(np.abs(z)))  # Two-tailed test

# Print results
print(f"Proportion for old layout: {p1:.4f}")
print(f"Proportion for new layout: {p2:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in conversion rates.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in conversion rates.")
```

```python
#Q.15:-15. A company is testing two different website layouts to see which one leads to higher click-through rates.

#Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.

#Use the following data:

#layout_a_clicks-128, 32, 33, 29, 31, 34, 30, 35, 36, 37]

#python

#layout_b_clicks (40, 41, 38, 42, 39, 44, 43, 41, 45, 47
#Ans:-
import numpy as np
from scipy import stats

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    """
    Perform A/B test analysis to compare click-through rates between two website layouts.

    Parameters:
    - layout_a_clicks (list or np.ndarray): Click-through rates for layout A.
    - layout_b_clicks (list or np.ndarray): Click-through rates for layout B.

    Returns:
    - t_statistic (float): The t-statistic of the test.
    - p_value (float): The p-value of the test.
    - df (float): Degrees of freedom for the test.
    """
    # Convert lists to numpy arrays
    layout_a = np.array(layout_a_clicks)
    layout_b = np.array(layout_b_clicks)

    # Calculate sample means and standard deviations
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    std_a = np.std(layout_a, ddof=1)
    std_b = np.std(layout_b, ddof=1)
    n_a = len(layout_a)
    n_b = len(layout_b)

    # Perform two-sample t-test
    t_statistic, p_value = stats.ttest_ind(layout_a, layout_b, equal_var=False)

    # Degrees of freedom using Welch-Satterthwaite equation for unequal variances
    df = ((std_a**2 / n_a + std_b**2 / n_b) ** 2) / \
         ((std_a**2 / n_a) ** 2 / (n_a - 1) + (std_b**2 / n_b) ** 2 / (n_b - 1))

    return t_statistic, p_value, df

# Example usage
layout_a_clicks = [128, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_statistic, p_value, df = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Print results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {df:.2f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in click-through rates between the two layouts.")
```

```python
#Q.18:- An HR department wants to investigate if there's a gender-based salary gap within the company. Develop

#a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees.

#Use the below code to generate synthetic data:

#python

#Generate synthetic salary data for male and female employees np.random.seed(0) # For reproducibility male_salaries = np.random.normal(loc=50000, scale=10000, size=20) female salaries np.random.normal(loc=55000, scale=9000, size=20)
#Ans:-
import numpy as np
from scipy import stats

def analyze_salary_gap(male_salaries, female_salaries):
    """
    Analyze if there is a statistically significant difference in average salaries
    between male and female employees.

    Parameters:
    - male_salaries (list or np.ndarray): Salaries of male employees.
    - female_salaries (list or np.ndarray): Salaries of female employees.

    Returns:
    - t_statistic (float): The t-statistic of the test.
    - p_value (float): The p-value of the test.
    - df (float): Degrees of freedom for the test.
    """
    # Convert lists to numpy arrays
    males = np.array(male_salaries)
    females = np.array(female_salaries)

    # Calculate sample means and standard deviations
    mean_males = np.mean(males)
    mean_females = np.mean(females)
    std_males = np.std(males, ddof=1)
    std_females = np.std(females, ddof=1)
    n_males = len(males)
    n_females = len(females)

    # Perform two-sample t-test
    t_statistic, p_value = stats.ttest_ind(males, females, equal_var=False)

    # Degrees of freedom using Welch-Satterthwaite equation for unequal variances
    df = ((std_males**2 / n_males + std_females**2 / n_females) ** 2) / \
         ((std_males**2 / n_males) ** 2 / (n_males - 1) + (std_females**2 / n_females) ** 2 / (n_females - 1))

    return t_statistic, p_value, df

# Generate synthetic salary data
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform analysis
t_statistic, p_value, df = analyze_salary_gap(male_salaries, female_salaries)

# Print results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {df:.2f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in average salaries between male and female employees.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average salaries between male and female employees.")
```

```python
#Q.19:- A manufacturer produces two different versions of a product and wants to compare their quality scores.

#Create a Python function to analyze quality assessment data, calculate the t-statistic, and decide whether there's a significant difference in quality between the two versions.

#Use the following data:

#python versiont scores [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85] version2 scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 78, 78, 80, 81 87)
#Ans:-
import numpy as np
from scipy import stats

def analyze_quality_scores(version1_scores, version2_scores):
    """
    Analyze if there is a statistically significant difference in quality scores
    between two versions of a product.

    Parameters:
    - version1_scores (list or np.ndarray): Quality scores for version 1.
    - version2_scores (list or np.ndarray): Quality scores for version 2.

    Returns:
    - t_statistic (float): The t-statistic of the test.
    - p_value (float): The p-value of the test.
    - df (float): Degrees of freedom for the test.
    """
    # Convert lists to numpy arrays
    version1 = np.array(version1_scores)
    version2 = np.array(version2_scores)

    # Calculate sample means and standard deviations
    mean_version1 = np.mean(version1)
    mean_version2 = np.mean(version2)
    std_version1 = np.std(version1, ddof=1)
    std_version2 = np.std(version2, ddof=1)
    n_version1 = len(version1)
    n_version2 = len(version2)

    # Perform two-sample t-test
    t_statistic, p_value = stats.ttest_ind(version1, version2, equal_var=False)

    # Degrees of freedom using Welch-Satterthwaite equation for unequal variances
    df = ((std_version1**2 / n_version1 + std_version2**2 / n_version2) ** 2) / \
         ((std_version1**2 / n_version1) ** 2 / (n_version1 - 1) + (std_version2**2 / n_version2) ** 2 / (n_version2 - 1))

    return t_statistic, p_value, df

# Example data
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 78, 78, 80, 81, 87]

# Perform analysis
t_statistic, p_value, df = analyze_quality_scores(version1_scores, version2_scores)

# Print results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {df:.2f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in quality scores between the two versions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in quality scores between the two versions.")
```

```python
#Q.20:- A restaurant chain collects customer satisfaction scores for two different branches. Write a program to analyze the scores, calculate the t-statistic, and determine if there's a statistically significant difference in customer satisfaction between the branches.

#Use the below data of scores:

#python

#branch a scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 54, 34, 5,435,4,4,5,345,4)

#branch_b_scores = [3, 4, 2, 3, 4, 3.4.2.3.3.4 3,3234,4,3,234,32,4,3,3,4,2,3,4,3]
#Ans:-
import numpy as np
from scipy import stats

def analyze_satisfaction_scores(branch_a_scores, branch_b_scores):
    """
    Analyze if there is a statistically significant difference in customer satisfaction scores
    between two branches.

    Parameters:
    - branch_a_scores (list or np.ndarray): Customer satisfaction scores for branch A.
    - branch_b_scores (list or np.ndarray): Customer satisfaction scores for branch B.

    Returns:
    - t_statistic (float): The t-statistic of the test.
    - p_value (float): The p-value of the test.
    - df (float): Degrees of freedom for the test.
    """
    # Convert lists to numpy arrays
    branch_a = np.array(branch_a_scores)
    branch_b = np.array(branch_b_scores)

    # Perform two-sample t-test
    t_statistic, p_value = stats.ttest_ind(branch_a, branch_b, equal_var=False)

    # Degrees of freedom using Welch-Satterthwaite equation for unequal variances
    std_a = np.std(branch_a, ddof=1)
    std_b = np.std(branch_b, ddof=1)
    n_a = len(branch_a)
    n_b = len(branch_b)
    
    df = ((std_a**2 / n_a + std_b**2 / n_b) ** 2) / \
         ((std_a**2 / n_a) ** 2 / (n_a - 1) + (std_b**2 / n_b) ** 2 / (n_b - 1))

    return t_statistic, p_value, df

# Example data
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 4, 4, 5, 4, 4, 5, 4, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 4, 4, 3, 4, 3, 3, 4, 2, 3, 4, 3]

# Perform analysis
t_statistic, p_value, df = analyze_satisfaction_scores(branch_a_scores, branch_b_scores)

# Print results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {df:.2f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in customer satisfaction scores between the branches.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in customer satisfaction scores between the branches.")
```

```python
#Q.21:- A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if

#there is a significant association between age groups and voter preferences.

##Use the below code to generate data:

#python

#np.random.seed(0)

#age groups np.random.choice(118-30, 31-50, 51, 51+1, size=30)

#voter preferences np.random.choice([Candidate A', 'Candidate B), size=30)
#Ans:-
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def perform_chi_square_test(age_groups, voter_preferences):
    """
    Perform a Chi-Square test of independence to determine if there is a significant association 
    between age groups and voter preferences.

    Parameters:
    - age_groups (list or np.ndarray): Age groups of voters.
    - voter_preferences (list or np.ndarray): Voter preferences (e.g., Candidate A or Candidate B).

    Returns:
    - chi2_statistic (float): The Chi-Square statistic.
    - p_value (float): The p-value of the test.
    - dof (int): Degrees of freedom.
    - expected (ndarray): Expected frequencies table.
    """
    # Create a DataFrame from the data
    data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

    # Create a contingency table
    contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

    # Perform Chi-Square test
    chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table)

    return chi2_statistic, p_value, dof, expected

# Generate synthetic data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51-70', '71+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Perform Chi-Square test
chi2_statistic, p_value, dof, expected = perform_chi_square_test(age_groups, voter_preferences)

# Print results
print(f"Chi-Square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table:\n{expected}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")
```

```python
#Q.22:-22. A company conducted a customer satisfaction survey to determine if there is a significant relationship between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a Chi- Square test to determine if there is a significant relationship between product satisfaction levels and customer regions.

#Sample data:

#python

#Sample data: Product satisfaction levels (rows) vs. Customer regions (columns) data = np.array([[50, 30, 40, 20), [30, 40, 30, 50), [20, 30, 40, 30]])
#Ans:-
import numpy as np
from scipy.stats import chi2_contingency

def chi_square_test(contingency_table):
    """
    Perform a Chi-Square test of independence to determine if there is a significant association 
    between product satisfaction levels and customer regions.

    Parameters:
    - contingency_table (np.ndarray): The contingency table summarizing product satisfaction levels vs. customer regions.

    Returns:
    - chi2_statistic (float): The Chi-Square statistic.
    - p_value (float): The p-value of the test.
    - dof (int): Degrees of freedom.
    - expected (np.ndarray): Expected frequencies table.
    """
    # Perform Chi-Square test
    chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table)

    return chi2_statistic, p_value, dof, expected

# Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
data = np.array([[50, 30, 40, 20],
                 [30, 40, 30, 50],
                 [20, 30, 40, 30]])

# Perform Chi-Square test
chi2_statistic, p_value, dof, expected = chi_square_test(data)

# Print results
print(f"Chi-Square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table:\n{expected}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant relationship between product satisfaction levels and customer regions.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between product satisfaction levels and customer regions.")
```

```python
#Q.24:- A company produces three different versions of a product: Standard, Premium, and Deluxe. The company wants to determine if there is a significant difference in customer satisfaction scores among the three product versions. They conducted a survey and collected customer satisfaction scores for each version from a random sample of customers. Perform an ANOVA test to determine if there is a significant difference in customer satisfaction scores.
#Use the following data:
#python
#Sample data: Customer satisfaction scores for each product version
#standard_scores = 180, 85, 90, 78, 88, 82, 92, 78, 85, 87) premium scores 190, 92, 88, 92, 95, 91, 96, 93, 89, 93] deluxe scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99)
#Ans:-
import numpy as np
from scipy.stats import f_oneway

def perform_anova(*args):
    """
    Perform a one-way ANOVA test to determine if there is a significant difference
    in customer satisfaction scores among different product versions.

    Parameters:
    - *args: Multiple arrays of customer satisfaction scores for different product versions.

    Returns:
    - f_statistic (float): The F-statistic of the ANOVA test.
    - p_value (float): The p-value of the test.
    """
    # Perform one-way ANOVA test
    f_statistic, p_value = f_oneway(*args)

    return f_statistic, p_value

# Sample data: Customer satisfaction scores for each product version
standard_scores = [180, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [190, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_statistic, p_value = perform_anova(standard_scores, premium_scores, deluxe_scores)

# Print results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in customer satisfaction scores among the product versions.")
```
